```python
#data = load_data.load_from_box()
data = load_data.load_from_file()
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,5))
fig = sns.lineplot(data=[data['A'], data['B'], data['C']],
                   dashes = False, markers = [".",".","."])
fig.set(xlabel="Date",ylabel="Blood Pressure (mm Hg)")
plt.title("Fake Blood Pressures Over Time", fontsize = 20)
plt.xticks(rotation=45,horizontalalignment='right')
plt.show()
```

```python
import pandas as pd
fig = sns.boxplot(x="variable", y="value", data=pd.melt(data))
fig.set(xlabel="Subject",ylabel="Blood Pressure (mm Hg)", title="Fake Blood Pressures")

plt.show()
```
